```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, color
```

```python
image = io.imread("images/wheat.jpg")
plt.imshow(image)
```

```python
R = image[:, :, 0]
G = image[:, :, 1]
B = image[:, :, 2]

# Convert to CIE Lab color space
lab_img = color.rgb2lab(image)
L = lab_img[:, :, 0]  # Lightness
a = lab_img[:, :, 1]  # Green-Red
b = lab_img[:, :, 2]  # Blue-Yellow

# Display all channels
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# RGB channels
axes[0, 0].imshow(R, cmap='Reds')
axes[0, 0].set_title('Red Channel')
axes[0, 0].axis('off')

axes[0, 1].imshow(G, cmap='Greens')
axes[0, 1].set_title('Green Channel')
axes[0, 1].axis('off')

axes[0, 2].imshow(B, cmap='Blues')
axes[0, 2].set_title('Blue Channel')
axes[0, 2].axis('off')

# Lab channels
axes[1, 0].imshow(L, cmap='gray')
axes[1, 0].set_title('L Channel (Lightness)')
axes[1, 0].axis('off')

axes[1, 1].imshow(a, cmap='RdYlGn_r')
axes[1, 1].set_title('a Channel (Green-Red)')
axes[1, 1].axis('off')

axes[1, 2].imshow(b, cmap='RdYlBu_r')
axes[1, 2].set_title('b Channel (Blue-Yellow)')
axes[1, 2].axis('off')
```

```python
image = io.imread("images/bush.png")
plt.imshow(image)
```

```python
hsv_img = color.rgb2hsv(image)

# Extract HSV channels
H = hsv_img[:, :, 0]  # Hue [0, 1]
S = hsv_img[:, :, 1]  # Saturation [0, 1]
V = hsv_img[:, :, 2]  # Value [0, 1]

# Display original image
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

axes[0, 0].imshow(image)
axes[0, 0].set_title('Original Image (RGB)', fontsize=14, fontweight='bold')
axes[0, 0].axis('off')

# Display HSV channels individually
axes[0, 1].imshow(H, cmap='hsv')
axes[0, 1].set_title('Hue Channel', fontsize=12)
axes[0, 1].axis('off')

axes[1, 0].imshow(S, cmap='gray')
axes[1, 0].set_title('Saturation Channel', fontsize=12)
axes[1, 0].axis('off')

axes[1, 1].imshow(V, cmap='gray')
axes[1, 1].set_title('Value Channel', fontsize=12)
axes[1, 1].axis('off')
```

```python
hsv_modified = hsv_img.copy()
hsv_modified[:, :, 2] = 1.0  # Set all Value pixels to 1

# Convert back to RGB for display
rgb_modified = color.hsv2rgb(hsv_modified)

# Display comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].imshow(image)
axes[0].set_title('Original Image', fontsize=14, fontweight='bold')
axes[0].axis('off')

axes[1].imshow(rgb_modified)
axes[1].set_title('Image with Value = 1 (Full Brightness)', fontsize=14, fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('bush_value_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```
