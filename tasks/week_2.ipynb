```python
import matplotlib.pyplot as plt
from skimage import io, exposure
import numpy as np

filename = "images/airfield.tif"
airfield = io.imread(filename)
plt.imshow(airfield)
plt.show()
```

```python
hist, bins = np.histogram(airfield.ravel(), bins=256, range=(0, 256))
plt.plot(bins[:-1], hist)
plt.show()
```

```python
alpha = 1.5
beta = 30

adjusted = np.clip(alpha * airfield + beta, 0, 255).astype(np.uint8)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(airfield)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(adjusted)
axes[1].set_title(f'Adjusted (α={alpha}, β={beta})')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
equalized = exposure.equalize_hist(airfield)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(airfield, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')

axes[0, 1].imshow(equalized, cmap='gray')
axes[0, 1].set_title('Histogram Equalized')
axes[0, 1].axis('off')

axes[1, 0].hist(airfield.ravel(), bins=256, color='gray', alpha=0.7)
axes[1, 0].set_title('Original Histogram')
axes[1, 0].set_xlabel('Pixel Intensity')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].hist(equalized.ravel(), bins=256, color='blue', alpha=0.7)
axes[1, 1].set_title('Equalized Histogram')
axes[1, 1].set_xlabel('Pixel Intensity')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print(f"Original - Min: {airfield.min()}, Max: {airfield.max()}, Mean: {airfield.mean():.2f}")
print(f"Equalized - Min: {equalized.min():.3f}, Max: {equalized.max():.3f}, Mean: {equalized.mean():.3f}")
```
